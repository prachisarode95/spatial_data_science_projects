```python
# 🌐 Step 6: Run Streamlit + Tunnel using ngrok
import os
import threading
import time

# Launch Streamlit in a separate thread
def run_streamlit():
    os.system('streamlit run uhi_dashboard.py')

threading.Thread(target=run_streamlit).start()
time.sleep(10)

# Launch ngrok tunnel
!ngrok http 8501
```
